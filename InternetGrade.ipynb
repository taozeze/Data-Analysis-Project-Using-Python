Now, we will find out how many students top the class based on availability of internet. To showcase this, we will plot a swarm chart.

```python
ave = sum(data.G3)/float(len(data))

data2 = pd.DataFrame(data[data.G3>0])

data2['ave_line'] = ave
data2['average'] = ['above average' if i > ave else 'under average' for i in data2.G3]
sns.swarmplot(x='internet', y = 'G3', hue = 'average',data= data2,palette={'above average':'lime', 'under average': 'red'})
plt.legend(loc="best")
plt.show()
```

As we can see above, students with internet access score higher than students without internet access.
